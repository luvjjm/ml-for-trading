```python
import util
import pandas as pd

symbollist=['FAKE2']

start_date='2005-12-31'
end_date='2014-12-07'

idx = pd.date_range(start_date, end_date)
df_data=util.get_data(symbollist, idx)
df_data.fillna(method="ffill", inplace=True)

util.plot_data(df_data, title='Incomplete Data (with forward fill)')
```
